## Scoring Method

### Moving Average

```python
# Normalize future trend scores 
def score_trend(row, model, max_increase, max_decrease):
    if pd.isna(row[model]):
        return np.nan  # Handling the NA values for the initial period without enough data
    change = row['interest_level'] - row[model]
    if change > 0:
        return 50 + (change / max_increase) * 50     # Scale to 50-100 range
    elif change < 0:
        return 50 + (change / abs(max_decrease)) * 50  # Scale to 0-50 range
    else:
        return 50
```

```python
# Visualize trend score on a sample data for comparison
def sample_visualize(sample, method):
    plt.figure(figsize = (12, 8))
    temp = sample[(sample['company_id']==228) & (sample['topic_id']==110)] 
    # Plot original interest level VS trend score
    sns.lineplot(x='week', y='interest_level', data=temp, label = 'interest_level')
    sns.lineplot(x='week', y='trend_score', data=temp, label = f'{method} trend score')

    plt.title(f'Interest level x {method} trend score')
    plt.xlabel('Week')
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
def sma(df_sma, t=4):
    # Calculate the Simple Moving Average (SMA) for the interest level
    df_sma['sma'] = df_sma.groupby(['company_id', 'topic_id'])['interest_level'].transform(lambda x: x.rolling(window=t, min_periods=1).mean())
    
    # Calculate the maximum observed increase and decrease for normalization based on SMA
    max_increase = (df_sma['sma'] - df_sma['interest_level'].shift(t - 1)).max()
    max_decrease = (df_sma['sma'] - df_sma['interest_level'].shift(t - 1)).min()
    
    df_sma['trend_score'] = df_sma.apply(lambda row: score_trend(row, 'sma', max_increase, max_decrease), axis = 1)
    return df_sma
```

```python
# As we see the trend is more smooth
temp = sma(df).copy()
sample_visualize(temp, 'SMA')
```

### Weighted moving average

```python
def wma(df_wma, t=4):
    # Weights for MA
    weights = np.arange(1, t + 1)[::-1] 
    df_wma['wma'] = df_wma['interest_level'].rolling(window=t).apply(lambda x: np.dot(x, weights) / weights.sum(), raw=True)
    
    max_increase = (df_wma['wma'] - df_wma['interest_level'].shift(t - 1)).max()
    max_decrease = (df_wma['wma'] - df_wma['interest_level'].shift(t - 1)).min()
    
    df_wma['trend_score'] = df_wma.apply(lambda row: score_trend(row, 'wma', max_increase, max_decrease), axis = 1)
    return df_wma
```

```python
# As we see the trend is more smooth
temp = wma(df).copy()
sample_visualize(temp, 'WMA')
```

- Proposed method is less volatile, since beta is less than 1
- WMA's std is greater than the interest levels std which may be caused by the fact that WMA is much sensitive to the recent changes

```python
metrics_report(temp, 'WMA')
```

### Exponential Moving Average

```python
def ema(df_ema, span=4):
    # Calculate EMA using the pandas ewm()
    df_ema['ema'] = df_ema.groupby(['company_id', 'topic_id'])['interest_level'].transform(lambda x: x.ewm(span=span, adjust=False).mean())
    
    max_increase = (df_ema['ema'] - df_ema['interest_level'].shift(span - 1)).max()
    max_decrease = (df_ema['ema'] - df_ema['interest_level'].shift(span - 1)).min()

    df_ema['trend_score'] = df_ema.apply(lambda row: score_trend(row, 'ema', max_increase, max_decrease), axis = 1)
    return df_ema
```

### Moving Median

```python
def median(df_med, t=4):
    weights = np.arange(1, t + 1)[::-1] 
    df_med['med'] = df_med.groupby(['company_id', 'topic_id'])['interest_level'].transform(lambda x: x.rolling(window=t, min_periods=1).median())
    max_increase = (df_med['med'] - df_med['interest_level'].shift(t - 1)).max()
    max_decrease = (df_med['med'] - df_med['interest_level'].shift(t - 1)).min()
    df_med['trend_score'] = df_med.apply(lambda row: score_trend(row, 'med', max_increase, max_decrease), axis = 1)
    return df_med
```

### Fourier analysis

```python
def fourier_ts(fourier_df, n=5):
    fourier_df['trend_score'] = np.nan
    
    for (company_id, topic_id), group in fourier_df.groupby(['company_id', 'topic_id']):
        if len(group) > 2 * n :  

            fft_values = np.fft.fft(group['interest_level'].values)
            fft_values[n:-n] = 0
            smoothed_values = np.real(np.fft.ifft(fft_values))
            
            fourier_df.loc[group.index, 'smoothed_values'] = smoothed_values
            
            trend_scores = group['interest_level'].values - smoothed_values
            max_increase = np.max(trend_scores) if np.max(trend_scores) != 0 else 1  # Avoid division by zero
            max_decrease = np.min(trend_scores) if np.min(trend_scores) != 0 else -1  # Avoid division by zero
            
            for index in group.index:
                fourier_df.at[index, 'trend_score'] = score_trend(fourier_df.loc[index], 'smoothed_values', max_increase, max_decrease)
        else:
            # Handle the case with insufficient data
            fourier_df.loc[group.index, 'trend_score'] = 50 
                
    return fourier_df
```

### Kalman Filter

```python
from filterpy.kalman import KalmanFilter

def score_trend_kf(interest_level, filtered_state, max_increase, max_decrease):
    if pd.isna(filtered_state):
        return np.nan
    change = interest_level - filtered_state
    if change > 0:
        return 50 + (change / max_increase if max_increase != 0 else 0) * 50
    elif change < 0:
        return 50 + (change / abs(max_decrease) if max_decrease != 0 else 0) * 50
    else:
        return 50
    

def kalman_filter(group):
    # Ensure the group is not empty and has no NaN values
    if not group['interest_level'].dropna().empty:
        # Initialize the Kalman filter
        kf = KalmanFilter(dim_x=1, dim_z=1)
        
        # Define the initial state and covariance
        kf.x = np.array([[0.]])  # initial state
        kf.P = np.array([[1.]])  # initial state covariance
        kf.R = 72 # Measurement noise
        kf.Q = 1e-7 # Process noise
        kf.F = np.array([[1.]])  # State transition matrix
        kf.H = np.array([[1.]])  # Measurement function
        
        # Apply the Kalman filter to the interest levels
        filtered_state = np.zeros_like(group['interest_level'].values)
        for i, val in enumerate(group['interest_level'].values):
            kf.predict()
            kf.update(np.array([[val]]))
            filtered_state[i] = kf.x.flatten()[0]
        
        group['filtered_state'] = filtered_state
    return group

def kalman_ts(kalman_df):
    kalman_df = kalman_df.groupby(['company_id', 'topic_id']).apply(kalman_filter, include_groups=False)
    
    kalman_df['difference'] = kalman_df['filtered_state'] - kalman_df['interest_level']
    kalman_df['max_increase'] = kalman_df.groupby(['company_id', 'topic_id'])['difference'].transform('max')
    kalman_df['max_decrease'] = kalman_df.groupby(['company_id', 'topic_id'])['difference'].transform('min')

    # Normalize the trend scores using the score_trend function in a vectorized way
    grouped = kalman_df.groupby(['company_id', 'topic_id'])
    
    kalman_df['trend_score'] = np.vectorize(score_trend_kf)(\
                            kalman_df['interest_level'], kalman_df['filtered_state'], kalman_df['max_increase'], kalman_df['max_decrease'])
    return kalman_df
```

```python
def tune_kalman(kalman_df, R, Q_params):
    std_list = []
    
    for Q in Q_params:
        kalman_df = kalman_df.groupby(['company_id', 'topic_id']).apply(lambda group: kalman_filter(group, R, Q), include_groups=False)
        kalman_df['difference'] = kalman_df['filtered_state'] - kalman_df['interest_level']
        kalman_df['max_increase'] = kalman_df.groupby(['company_id', 'topic_id'])['difference'].transform('max')
        kalman_df['max_decrease'] = kalman_df.groupby(['company_id', 'topic_id'])['difference'].transform('min')

        # Normalize the trend scores using the score_trend function in a vectorized way
        grouped = kalman_df.groupby(['company_id', 'topic_id'])

        kalman_df['trend_score'] = np.vectorize(score_trend_kf)(\
                                kalman_df['interest_level'], kalman_df['filtered_state'], kalman_df['max_increase'], kalman_df['max_decrease'])

        std = kalman_df.groupby(['company_id', 'topic_id'])['trend_score'].std().median()
        std_list.append(std)
    return std_list
```

## Evaluation

We are gonna use std and customer "beta" metric for evaluation
Inspired by stock market intuition we want to assess volatility
- Beta > 1: Companies with a Beta greater than 1 are more volatile than the market trend. Their trend scores fluctuate more in response to the overall market.
- Beta < 1: Companies with a Beta less than 1 are less volatile than the market trend. Their trend scores are relatively stable even when the market trend scores fluctuate.
- Beta ≈ 1: Companies with a Beta around 1 move in tandem with the market. 

```python
def metrics_report(sample, method):
    sample['market_trend_score'] = sample.groupby(['company_id', 'topic_id'])['interest_level'].transform(lambda x: x.rolling(window=4, min_periods=1).mean())
    sample = sample.dropna(subset=['trend_score', 'market_trend_score'])
    # Covariance between company's trend scores and market trend scores
    cov_mat = np.cov(sample['trend_score'], sample['market_trend_score'])
    covariance = cov_mat[0, 1]

    market_variance = sample['market_trend_score'].var()

    # Trend Beta
    beta = covariance / market_variance
    
    std = sample.groupby(['company_id', 'topic_id'])['interest_level'].std().mean()
    std2 = sample.groupby(['company_id', 'topic_id'])['trend_score'].std().mean()
    
    print(f'Beta value for the {method} method: {beta}'\
          f'\nstd of interest level: {std}'\
          f'\nstd of trend score: {std2}')
```

## Are there any groups of keywords that are explored together?

```python
# Grouping topics explored together by the same company in the same week
grouped_topics = df.groupby(['company_id', 'week'])['topic_id'].apply(list)

co_occurrence = {}

for topics in grouped_topics:
    # Avoid double counting (A,B) (B, A)
    unique_pairs = {tuple(sorted([topics[i], topics[j]])) for i in range(len(topics)) for j in range(i + 1, len(topics))}
    for pair in unique_pairs:
        if pair in co_occurrence:
            co_occurrence[pair] += 1
        else:
            co_occurrence[pair] = 1

co_occurrence_df = pd.DataFrame(list(co_occurrence.items()), columns=['Topic_Pair', 'Co-occurrence_Count'])
co_occurrence_df.sort_values(by='Co-occurrence_Count', ascending=False, inplace=True)
```
